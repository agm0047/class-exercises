```python
import pandas as pd
import geopandas as gpd
import shapely
from shapely.geometry import Point
```

```python
zipcodes= pd.read_csv(r'C:\Users\agmon\Documents\408-HW2\2025_Gaz_zcta_national.txt',sep= '|', dtype={'GEOID' : str})
```

```python
df=zipcodes
```

```python
zipcodes
```

```python
geometry = [Point(xy) for xy in zip(df["INTPTLONG"] , df['INTPTLAT'])]

```

```python
gdf= gpd.GeoDataFrame(df, geometry=geometry,columns=['GEOID','INTPTLAT', 'INTPTLONG'], crs ="EPSG:4326")
```

```python
gdf.to_file("zipcodes_with_points.shp")
```
